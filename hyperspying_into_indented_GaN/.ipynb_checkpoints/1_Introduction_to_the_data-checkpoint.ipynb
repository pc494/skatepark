This notebook contains the first introduction to this dataset. The data is loaded, plotted, and a peakfinder is chosen. One of the parameters for peak finding is calibrated. It should be noted that ** We do not adjust for the distortion ** - although this will be performed on our next sample.

```python
import hyperspy.api as hs
from matplotlib import pyplot as plt
import numpy as np

%matplotlib inline
```

```python
signal_zero = hs.load("/home/pc494/Documents/data/GaN_intro/GaN_indent.blo")
if False:
    signal_half = hs.load("/home/pc494/Documents/data/GaN_intro/GaN_indent_0-5degree.blo")
    signal_full = hs.load("/home/pc494/Documents/data/GaN_intro/GaN_indent_1degree.blo")
```

```python
signal_zero
```

```python
image_data = signal_zero.inav[250].inav[250].data
plt.imshow(image_data,cmap='viridis'); ## We are at the (real space) middle of the pattern
```

```python
import pycrystem as pc
peak_finder_one = pc.utils.peakfinders2D.find_peaks_minmax 
```

```python
thresholds = np.linspace(15,100,9)
peak_count = np.zeros_like(thresholds)
for index,thresh in enumerate(thresholds):
    peak_count[index] = peak_finder_one(image_data,threshold=thresh).shape[0]
```

```python
plt.scatter(thresholds,peak_count);
plt.xlabel('Threshold Value');
plt.ylabel('Peak Count');
```

```python
peaks = peak_finder_one(image_data,threshold=thresholds[1])
peak_index = np.floor(peaks).astype(int);
plt.imshow(image_data,cmap='viridis')
plt.scatter(peak_index.T[1],peak_index.T[0],marker='+',color='r',s=30)
```
